```python
import json
import pandas as pd

# Load your GeoJSON file
with open("historic_adt.json") as f:
    data = json.load(f)

# Extract the "features" list
features = data["features"]

# Flatten the JSON into a table-like structure
rows = []
for feature in features:
    props = feature["properties"]
    coords = feature["geometry"]["coordinates"]
    start_coord = coords[0]
    end_coord = coords[-1]
    
    # Assuming coordinates are in [longitude, latitude] order (standard GeoJSON format)
    rows.append({
        "segment_id": props.get("segment_id"),
        "year": props.get("year"),
        "aadt": props.get("aadt"),
        "start_lon": start_coord[0],
        "start_lat": start_coord[1],
        "end_lon": end_coord[0],
        "end_lat": end_coord[1],
        "num_points": len(coords)
    })

# Convert to DataFrame
df = pd.DataFrame(rows)

# Preview or save to CSV
# print(df.head())
# df.to_csv("output_table.csv", index=False)
```

```python
df
```

```python
green_max_lat = 40.110620
green_max_lon = 88.219279
green_min_lat = 40.110193
green_min_lon = -88.235437
```

```python
def overlaps_green(row):
    # Check if either start or end is in the latitude band
    lat_ok = ((row["start_lat"] >= green_min_lat) and (row["start_lat"] <= green_max_lat)) or ((row["end_lat"] >= green_min_lat) and (row["end_lat"] <= green_max_lat))
    # Check if the segment’s longitudes overlap the corridor
    seg_lon_min = min(row["start_lon"], row["end_lon"])
    seg_lon_max = max(row["start_lon"], row["end_lon"])
    lon_overlap = not (seg_lon_max < green_min_lon or seg_lon_min > green_max_lon)
    return lat_ok & lon_overlap

df["is_on_green"] = df.apply(overlaps_green, axis=1)

# Then filter
df[df.is_on_green == True]
```
