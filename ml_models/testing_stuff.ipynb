```python
import torch
import torch.nn as nn
```

```python
#preparing datasets
from sklearn.preprocessing import MinMaxScaler
features = ['pm10', 'carbon_monoxide', 'nitrogen_dioxide', 'sulphur_dioxide',
            'ozone', 'carbon_dioxide', 'aerosol_optical_depth', 'methane',
            'dust', 'uv_index', 'uv_index_clear_sky']

target = 'pm2_5'

# Scale features and target
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(df[features])
y_scaled = scaler_y.fit_transform(df[[target]])
```

```python
import numpy as np
```

```python
def create_sequences(X, y, seq_length=7):
    xs, ys = [], []
    for i in range(seq_length, len(X)):
        xs.append(X[i-seq_length:i])
        ys.append(y[i])
    return np.array(xs), np.array(ys)

seq_length = 24  # past 7 days
X_seq, y_seq = create_sequences(X_scaled, y_scaled, seq_length)

# Train-test split (chronological)
train_size = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:train_size], X_seq[train_size:]
y_train, y_test = y_seq[:train_size], y_seq[train_size:]

# Convert to PyTorch tensors
X_train = torch.from_numpy(X_train).float()
y_train = torch.from_numpy(y_train).float()
X_test = torch.from_numpy(X_test).float()
y_test = torch.from_numpy(y_test).float()
```

```python
class LSTM1(nn.Module):
    def __init__(self, n_features):
        super().__init__()
        self.model = nn.LSTM(input_size=n_features, hidden_size = 64, num_layers = 1, batch_first=True)
        self.fc = nn.Linear(64, 1)
    
    def forward(self, x):
        out, _ = self.model(x)
        out = self.fc(out[:, -1, :])  # use last time step output
        return out.squeeze()
```

```python
n_features = X_train.shape[2]
model_2 = LSTM1(n_features)
```

```python
prev_loss = 10000000
from torch import optim
loss_fn = nn.MSELoss()
optimizer = optim.Adam(params = model_2.parameters(), lr = 0.001)
for epoch in range(10000):
    model_2.train()
    optimizer.zero_grad()
    y_pred = model_2(X_train)
    loss = loss_fn(y_pred, y_train)
    if (loss.item() > prev_loss):
        break
    else:
        prev_loss = loss.item()
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 10 == 0:
        model_2.eval()
        with torch.no_grad():
            val_loss = loss_fn(model_2(X_test), y_test)
        print(f"Epoch {epoch+1}, Train Loss: {loss.item():.4f}, Val Loss: {val_loss.item():.4f}")
```
