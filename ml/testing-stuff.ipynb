```python
import pandas as pd
import sqlite3 as sql
```

```python
con = sql.connect('../data_collection/campus_data.db')
cur = con.cursor()
df_aq = pd.read_sql_query("SELECT * FROM air_quality", con)
```

```python
df_aq
```

```python
import matplotlib.pyplot as plt
df_aq['timestamp'] = pd.to_datetime(df_aq['datetime_utc'])
plt.plot(df_aq['timestamp'], df_aq['value'])
```

```python
df_weather = pd.read_sql_query("SELECT * FROM weather", con)
df_weather
```

```python
df_weather['timestamp'] = pd.to_datetime(df_weather['timestamp'])
df_weather['wind'] = pd.to_numeric(df_weather['wind'].str[:-4])
plt.plot(df_weather['timestamp'], df_weather['wind'])
```
